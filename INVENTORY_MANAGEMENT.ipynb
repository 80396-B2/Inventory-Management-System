# INVENTORY MANAGEMENT PROJECT

```python
import json
import random as rd
from datetime import datetime
import time
fd = open("inventory.json",'r')
txt= fd.read()
fd.close()
inv= json.loads(txt)
codelist=[]
quanlist=[] 
amtlist=[]
name=input("Enter your name: ")


def stockavailable():
    print('=============================')
    print('= List Of Drugs  =')
    print('=============================')
    print('{:<11} {:<17} {:>8}'.format("Item code","Drug_name","Power(in mg)"))
    print('===========================================')
    for code in inv.keys():
        print('{:<11} {:<17} {:<8}'.format(str(code),str(inv.get(code, {}).get('Drug_name')),str(inv.get(code, {}).get('mg-'))))
    print('===========================================')
    askfun()

def askfun():
    while True:
        try:
            itemcode=int(input("Please enter the item code of the required medicines:  "))
        except:
            print("\nIncorrect item code!! Please try again!!")
            continue
        else:
            if itemcode in [*range(11101,11131)]:
                break
            else:
                print("\nIncorrect item code!! Please try again!!")
    while True:
        try:
            quant=int(input("Enter quantity:  "))
        except:
            print("Looks like you did not enter a number!! Please try again!!")
            continue
        else:
            break
    print('===========================================')
    strin=".................. "
    print("Checking for Availability and permission for sale",end=' ')
    for i in strin:
        print(i,end='')
        time.sleep(0.2)
    print('\n===========================================')
    check(str(itemcode),quant)
def check(itemcode,quant):
    if inv.get(itemcode, {}).get('Over_the_counter') =="Yes":
        if int(inv.get(itemcode, {}).get('Quantity_avilable'))>=quant:
            print("The item is available in stock and is a non prescription drug")
            print("********Item was successfully added to cart********")
            inv[itemcode]["Quantity_avilable"]-=quant
            codelist.append(itemcode)
            quanlist.append(quant)
        elif int(inv.get(itemcode, {}).get('Quantity_avilable'))<quant:
            print("SORRY!! we have only ",inv.get(itemcode, {}).get('Quantity_avilable')," units available")
            print("Do you still wanna proceed??")
            while True:
                try:
                    print('===========================================')
                    cho=int(input("Press 1 for Yes and 2 for No: "))
                except ValueError:
                    print("Incorrect Choice Try Again!!")
                    continue
                else:
                    if cho==1:
                        print("********Item was successfully added to cart********")
                        tempitm=int(inv.get(itemcode, {}).get('Quantity_avilable'))
                        inv[itemcode]["Quantity_avilable"]-=inv[itemcode]["Quantity_avilable"]
                        codelist.append(itemcode)
                        quanlist.append(tempitm)
                        break
                    elif cho==2:
                        break
                    else:
                        print("Incorrect Choice Try Again!!")
              
            
    else:
        print("The item is not sold without a prescription!!")       
    
    while True:
        try:
            print('===========================================')
            press=int(input("To add more items to cart PRESS 1 or to checkout PRESS 2: "))
        except ValueError:
            print("Looks like you did not enter an integer!")
            continue
        else:
            if press==1:
                askfun()
                break
            elif press==2:
                bill(codelist,quanlist)
                break
            else:
                print("Incorrect Choice Try Again!!")
            
def bill(codelist,quanlist):
    cnt=1
    sums=0
    print("\n")
    print('{:>44}'.format("Bill"))
    print("----------------------------------------------------------------------------------------------")
    print('{:<1}{:<70}{:<1} {:<1}'.format("NAME:",name,"BILL NO:",rd.randint(564264,987431)))
    dto = datetime.now()
    print('{:<1}{:<1}{:<1}{:<1}{:<1}{:<69}{:<1}{:<1}{:<1}{:<1}{:<1}{:<1}'.format("DATE:",dto.day,"-",dto.month,"-",dto.year,"TIME: ",dto.hour,":",dto.minute,":",dto.second))
    print("----------------------------------------------------------------------------------------------")
    print('{:<10}{:<25}{:<10}{:<15}{:<15}{:<15}'.format("Sr No","Item Name","Qty","Exp","Barcode","Amount"))
    print("----------------------------------------------------------------------------------------------")
    for cdn,qty in zip(codelist,quanlist):
        print('{:<10}{:<25}{:<10}{:<15}{:<15}{:<15}'.format(cnt,inv.get(cdn, {}).get('Drug_name'),qty,inv.get(cdn, {}).get('Expiry'),inv.get(cdn, {}).get('Bar_code_no'),inv.get(cdn, {}).get('Price')))
        amtlist.append(inv.get(cdn, {}).get('Price'))
        cnt+=1
    print("----------------------------------------------------------------------------------------------")
    for i,j in zip(quanlist,amtlist):
        sums+=i*j
    print('{:<1}{:<70}{:<1}{:<1} {:<1}'.format("MEMO:","Online Payment","NETAMT:","Rs",sums))
    
stockavailable()    
```
